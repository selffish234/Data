```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
def bin_dist(k, n, p):
    nck = factorial(n) / (factorial(k) * factorial(n-k))
    pd = nck * p**k * (1-p)**(n-k)
    return pd
```

```python
x = np.arange(16)
pd1 = np.array([bin_dist(k, 15, 0.3) for k in range(16)])
plt.ylim(0, 0.3)
plt.text(12.5, 0.28, 'n, p = 15, 0.3')
plt.bar(x, pd1, color = 'lightcoral')
plt.show()
```

```python
x = np.arange(16)
pd2 = np.array([bin_dist(k, 15, 0.5) for k in range(16)])
plt.ylim(0, 0.3, 0.1)
plt.text(12.5, 0.28, 'n, p = 15, 0.5')
plt.bar(x, pd2, color = 'turquoise')
plt.show()
```

```python
x = np.arange(16)
pd3 = np.array([bin_dist(k, 15, 0.7) for k in range(16)])
plt.ylim(0, 0.3, 0.1)
plt.text(12.5, 0.28, 'n, p = 15, 0.7')
plt.bar(x, pd3, color = 'blue')
plt.show()
```

# subplot

```python
x = np.arange(16)
pd1 = np.array([bin_dist(k, 15, 0.3) for k in range(16)])
plt.subplot(3, 1, 1)
plt.ylim(0, 0.3, 0.1)
plt.text(12.5, 0.25, 'n, p = 15, 0.3')
plt.bar(x, pd1, color = 'lightcoral')

pd2 = np.array([bin_dist(k, 15, 0.5) for k in range(16)])
plt.subplot(3, 1, 2)
plt.ylim(0, 0.3, 0.1)
plt.text(12.5, 0.25, 'n, p = 15, 0.5')
plt.bar(x, pd2, color = 'turquoise')

pd3 = np.array([bin_dist(k, 15, 0.7) for k in range(16)])
plt.subplot(3, 1, 3)
plt.ylim(0, 0.3, 0.1)
plt.text(12.5, 0.25, 'n, p = 15, 0.7')
plt.bar(x, pd3, color = 'blue')
plt.show()
```

# random.binomial

```python
np.random.seed(0)
n = 15
p = 0.3
a = np.random.binomial(n, p, 1000)
plt.xlim(0, 16)
plt.ylim(0, 400)
plt.text(12.5, 270, 'n, p = 15, 0.3')
plt.hist(a, color = 'lightcoral')
plt.show()
```
