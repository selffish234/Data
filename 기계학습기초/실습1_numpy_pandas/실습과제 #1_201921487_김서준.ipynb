# numpy basic

```python
# numpy를 사용하여 다음 list를 2X2 행렬로 만드시오.
ex_list = [[1,2,3,4]]
np.array(ex_list).reshape(2,2)#numpy를 사용하여 ex_list를 2X2 행렬로 만듦
```

```python
# numpy를 사용하여 다음 list를 1차원 행렬로 만드시오.
ex_list = [[1,2,3,4]]
np.array(ex_list).reshape(-1) #reshape에 -1을 넣으면 1차원 행렬로 바꾸어 줌
```

```python
# numpy를 사용하여 다음 두 list를 합치고 평균과 합울 구하시오
ex_list_1 = np.array([i for i in range(0,100,2)])
ex_list_2 = np.array([i for i in range(100,200,2)])
a= ex_list_1
b = ex_list_2
c = a+b
print("합:",c.sum(),"평균:",c.mean())#객체.sum():합 구하는 메소드, 객체.mean() 평균구하는 메소드
```

```python
#seaborn 혹은 plt를 활용하여 산포도 (scatter plot)를 그리시오.
# x축은 age , y축은 fare
titanic_df = sns.load_dataset('titanic')
 
sns.set_style('whitegrid')
sns.scatterplot(data=titanic_df, x='age', y='fare')
```

```python
#seaborn 혹은 plt를 활용하여 산포도 (line plot)를 그리시오.
# x축은 age , y축은 fare
titanic_df = sns.load_dataset('titanic')
 
sns.set_style('whitegrid')
sns.lineplot(data=titanic_df, x='age', y='fare')
```

```python
#seaborn 혹은 plt를 활용하여 "who" column의 히스토그램을 그리시오.
titanic_df = sns.load_dataset('titanic')
sns.set_style('whitegrid')
sns.histplot(x=titanic_df['who'])
```
